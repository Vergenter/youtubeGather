```python
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
import numpy as np
import os
DEVELOPER_KEY = os.environ['YOUTUBE_API_KEY_V3']
YOUTUBE_API_SERVICE_NAME = 'youtube'
YOUTUBE_API_VERSION = 'v3'
youtube = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION,developerKey=DEVELOPER_KEY)
```

```python
user = "UCOhx1Dx674JbX4vK-QHGn7g"
request = youtube.channels().list(part='contentDetails',id=user)
response = request.execute()
```

```python
uploads = response["items"][0]["contentDetails"]["relatedPlaylists"]["uploads"]
```

```python
def readResultToEnd(makeRequest):
    accumulatedResult = np.array([])
    pageToken=""
    while True:
        response = makeRequest(pageToken).execute()
        accumulatedResult = np.append(accumulatedResult,response["items"])
        if "nextPageToken" not in response:
            break
        pageToken=response["nextPageToken"]
    return accumulatedResult
        
```

```python
results = readResultToEnd(lambda pageToken: youtube.playlistItems().list(part="id,contentDetails",playlistId=uploads,maxResults=50,pageToken=pageToken))
```

```python
getVideo = lambda videoID: youtube.videos().list(part='snippet,contentDetails,recordingDetails,liveStreamingDetails,statistics,status,topicDetails',id=videoID).execute()
```

```python
videos = np.array(list(map(lambda result: getVideo(result['contentDetails']['videoId']),results)))
```

```python
videos[0]['items'][0]['statistics']["dislikeCount"]
```

```python
import dateutil.parser
def parseVideoToUsefullData(video):
    return {
        "title":video['items'][0]['snippet']['title'],
        "published":dateutil.parser.parse(video['items'][0]['snippet']['publishedAt']),
        "views":video['items'][0]['statistics']['viewCount'],
        "likes":video['items'][0]['statistics']['likeCount'],
        "dislikes":video['items'][0]['statistics']["dislikeCount"],
        "commentsCount":video['items'][0]['statistics']['commentCount'],
        "isStream":'liveStreamingDetails' in video['items'][0]
    }
```

```python
reducedVideos = list(map(parseVideoToUsefullData,videos))
```

```python
reducedVideos[0]
```

```python
from pandas import DataFrame,to_numeric
dt = DataFrame(reducedVideos)
dt['title'] = dt['title'].astype("string")
dt['views'] = to_numeric(dt['views'])
dt['dislikes'] = to_numeric(dt['dislikes'])
dt['likes'] = to_numeric(dt['likes'])
dt['commentsCount'] = to_numeric(dt['commentsCount'])
```
